```python
import sklearn as skl
import sklearn
from sklearn import datasets
from sklearn import decomposition
```

```python
digits = datasets.load_digits()
```

```python
M,N = 3,3
f,ax = plt.subplots(M, N)
for i in range(M):
    for j in range(N):
        ax[i,j].imshow(digits.images[i*N+j], cmap='Greys', interpolation='None')
        ax[i,j].set_title('Number %d' % digits.target[i*N+j])
        ax[i,j].get_xaxis().set_visible(False)
        ax[i,j].get_yaxis().set_visible(False)
```

```python
y = digits.target
```

```python
y.shape
```

```python
X = digits.data
pca = decomposition.PCA()
pca.fit(X)
var_ratios_true = pca.explained_variance_ratio_

plot(var_ratios_true,'r')

legend()
title('Explained variance ratio')
```

```python
plot( cumsum(var_ratios_true),'r' )
plot([0,64],[0.9,0.9],':b',label='90% of variation')
plot([0,64],[1,1],'--b',label='100% of variation')
legend()
title('Cumulative explained variance')
```

Incomplete PCA - with incomplete SVD (TruncatedSVD) - when need to get only first several components.

```python
pca = decomposition.PCA(n_components=2)
P =pca.fit_transform((X-X.mean(0)))
```

```python
P.shape
```

```python
scatter(P[:,0],P[:,1],c=digits.target)
```

# Iris dataset

```python
dataset = sklearn.datasets.load_iris()
```

```python
dataset.keys()
```

```python
X = dataset.data
Y = dataset.target
```

```python
pca = decomposition.PCA(n_components=2)
P =pca.fit_transform((X-X.mean(0))/X.std(0))
```

```python
scatter(P[:,0],P[:,1], c=Y)
```
